```python
import csv
import os
```

```python


class EmployeeManager:
    def __init__(self, filename='employees.csv'):
        self.filename = filename
        self.fieldnames = ['ID', 'Name', 'Position', 'Salary', 'Email']
        self.employees = self._load_data()

    def _load_data(self):
        employees = {}
        if not os.path.exists(self.filename):
            return employees
        
        with open(self.filename, mode='r', newline='', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            for row in reader:
                employees[row['ID']] = row
        return employees

    def _save_data(self):
        with open(self.filename, mode='w', newline='', encoding='utf-8') as file:
            writer = csv.DictWriter(file, fieldnames=self.fieldnames)
            writer.writeheader()
            writer.writerows(self.employees.values())

    def add_employee(self):
        print("\n---  Add New Employee ---")
        emp_id = input("Enter Employee ID: ").strip()
        if emp_id in self.employees:
            print(f" Error: Employee with ID '{emp_id}' already exists.")
            return

        name = input("Enter Name: ").strip()
        position = input("Enter Position: ").strip()
        email = input("Enter Email: ").strip()

        while True:
            try:
                salary = float(input("Enter Salary: ").strip())
                break
            except ValueError:
                print(" Invalid input. Please enter a number for the salary.")

        self.employees[emp_id] = {
            'ID': emp_id, 'Name': name, 'Position': position, 
            'Salary': salary, 'Email': email
        }
        self._save_data()
        print(f" Employee '{name}' was added successfully!")

    def view_all_employees(self):
        print("\n--- ðŸ‘¥ All Employees List ---")
        if not self.employees:
            print("No employee records found.")
            return

        print(f"{'ID':<5} {'Name':<20} {'Position':<20} {'Salary':<10} {'Email':<25}")
        print("-" * 85)

        for emp in self.employees.values():
            salary_val = float(emp['Salary'])
            print(f"{emp['ID']:<5} {emp['Name']:<20} {emp['Position']:<20} "
                  f"{salary_val:<10.2f} {emp['Email']:<25}")

    def update_employee(self):
        print("\n---  Update Employee ---")
        emp_id = input("Enter the employee's ID to update: ").strip()
        
        if emp_id not in self.employees:
            print(f"Error: No employee found with ID '{emp_id}'.")
            return

        employee = self.employees[emp_id]
        print("\nEnter new details (press Enter to skip and keep the current value):")

        new_name = input(f"New Name (current: {employee['Name']}): ").strip()
        if new_name:
            employee['Name'] = new_name

        new_position = input(f"New Position (current: {employee['Position']}): ").strip()
        if new_position:
            employee['Position'] = new_position

        new_email = input(f"New Email (current: {employee['Email']}): ").strip()
        if new_email:
            employee['Email'] = new_email

        while True:
            new_salary_str = input(f"New Salary (current: {employee['Salary']}): ").strip()
            if not new_salary_str:
                break
            try:
                employee['Salary'] = float(new_salary_str)
                break
            except ValueError:
                print(" Invalid input. Please enter a number for the salary.")

        self._save_data()
        print(f" Employee ID '{emp_id}' has been updated.")

    def delete_employee(self):
        print("\n--- ðŸ—‘ï¸ Delete Employee ---")
        emp_id = input("Enter the ID of the employee to delete: ").strip()

        if emp_id not in self.employees:
            print(f"Error: No employee found with ID '{emp_id}'.")
            return

        emp_name = self.employees[emp_id]['Name']
        confirm = input(f"Are you sure you want to delete {emp_name}? (yes/no): ").strip().lower()
        
        if confirm == 'yes':
            del self.employees[emp_id]
            self._save_data()
            print(f" Employee ID '{emp_id}' was deleted.")
        else:
            print("Deletion cancelled.")

    def search_employee(self):
        print("\n---  Search Employee ---")
        emp_id = input("Enter the employee ID to search for: ").strip()
        
        employee = self.employees.get(emp_id)

        if employee:
            print("\n--- Employee Found ---")
            print(f"ID:       {employee['ID']}")
            print(f"Name:     {employee['Name']}")
            print(f"Position: {employee['Position']}")
            print(f"Salary:   {float(employee['Salary']):.2f}")
            print(f"Email:    {employee['Email']}")
        else:
            print(f" No employee found with ID '{emp_id}'.")

def main():
    manager = EmployeeManager()

    while True:
        print("\n=====  Employee Management System =====")
        print("1. Add Employee")
        print("2. View All Employees")
        print("3. Update Employee")
        print("4. Delete Employee")
        print("5. Search Employee")
        print("6. Exit")
        print("========================================")

        choice = input("Enter your choice: ").strip()

        if choice == '1':
            manager.add_employee()
        elif choice == '2':
            manager.view_all_employees()
        elif choice == '3':
            manager.update_employee()
        elif choice == '4':
            manager.delete_employee()
        elif choice == '5':
            manager.search_employee()
        elif choice == '6':
            print("Goodbye! ")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 6.")

if __name__ == "__main__":
    main()
```
