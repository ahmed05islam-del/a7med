```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder, FunctionTransformer
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier


from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report, confusion_matrix


df = pd.read_csv('/content/train_u6lujuX_CVtuZ9i (1).csv')
```

```python
df = df.drop(columns=['Loan_ID'])
```

```python
sns.countplot(x='Loan_Status', data=df)
plt.title('Class Distribution: Approved (Y) vs Not Approved (N)')
plt.show()
```

```python
df['Dependents'] = df['Dependents'].replace(to_replace='3+', value='3')
```

```python
df['TotalIncome'] = df['ApplicantIncome'] + df['CoapplicantIncome']
```

```python

numeric_features = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'TotalIncome']
categorical_features = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area', 'Credit_History']

X = df.drop('Loan_Status', axis=1)
y_raw = df['Loan_Status']

le = LabelEncoder()
y = le.fit_transform(y_raw)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])


categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'))
])
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    try:
        roc = roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
    except:
        roc = roc_auc_score(y_test, y_pred)

    print(f"--- {model_name} Results ---")
    print(f"Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}, ROC-AUC: {roc:.4f}")
    print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
    return pd.DataFrame([[model_name, acc, prec, rec, f1, roc]],
                        columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1', 'ROC-AUC'])

results_df = pd.DataFrame(columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1', 'ROC-AUC'])
```

```python

lr_baseline = Pipeline(steps=[('preprocessor', preprocessor),
                              ('classifier', LogisticRegression(random_state=42, max_iter=1000))])
lr_baseline.fit(X_train, y_train)
results_df = pd.concat([results_df, evaluate_model(lr_baseline, X_test, y_test, 'LR Baseline (L2)')], ignore_index=True)
lr_l1 = Pipeline(steps=[('preprocessor', preprocessor),
                        ('classifier', LogisticRegression(penalty='l1', solver='liblinear', random_state=42, max_iter=1000))])
lr_l1.fit(X_train, y_train)
results_df = pd.concat([results_df, evaluate_model(lr_l1, X_test, y_test, 'LR Baseline (L1)')], ignore_index=True)
```

```python
dt_baseline_gini = Pipeline(steps=[('preprocessor', preprocessor),
                                   ('classifier', DecisionTreeClassifier(criterion='gini', random_state=42, max_depth=5))])
dt_baseline_gini.fit(X_train, y_train)
results_df = pd.concat([results_df, evaluate_model(dt_baseline_gini, X_test, y_test, 'DT Baseline (Gini, d=5)')], ignore_index=True)

dt_baseline_entropy = Pipeline(steps=[('preprocessor', preprocessor),
                                      ('classifier', DecisionTreeClassifier(criterion='entropy', random_state=42, max_depth=5))])
dt_baseline_entropy.fit(X_train, y_train)
results_df = pd.concat([results_df, evaluate_model(dt_baseline_entropy, X_test, y_test, 'DT Baseline (Entropy, d=5)')], ignore_index=True)


plt.figure(figsize=(20,10))
encoded_cat_names = dt_baseline_gini.named_steps['preprocessor'].named_transformers_['cat']['onehot'].get_feature_names_out(categorical_features)
all_feature_names = numeric_features + list(encoded_cat_names)
plot_tree(dt_baseline_gini.named_steps['classifier'], feature_names=all_feature_names, filled=True, class_names=['N', 'Y'], rounded=True)
plt.show()
```

```python
rf_baseline = Pipeline(steps=[('preprocessor', preprocessor),
                              ('classifier', RandomForestClassifier(random_state=42))])
rf_baseline.fit(X_train, y_train)
results_df = pd.concat([results_df, evaluate_model(rf_baseline, X_test, y_test, 'RF Baseline')], ignore_index=True)

ada_baseline = Pipeline(steps=[('preprocessor', preprocessor),
                               ('classifier', AdaBoostClassifier(random_state=42))])
ada_baseline.fit(X_train, y_train)
results_df = pd.concat([results_df, evaluate_model(ada_baseline, X_test, y_test, 'AdaBoost Baseline')], ignore_index=True)
```

```python
lr_pipe = Pipeline(steps=[('preprocessor', preprocessor),
                          ('classifier', LogisticRegression(solver='liblinear', random_state=42, max_iter=5000))])

param_grid_lr = {
    'classifier__penalty': ['l1', 'l2'],
    'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100]
}

grid_lr = GridSearchCV(lr_pipe, param_grid_lr, cv=5, scoring='f1', n_jobs=-1)
grid_lr.fit(X_train, y_train)

print(f"Best LR Parameters: {grid_lr.best_params_}")
best_lr = grid_lr.best_estimator_
results_df = pd.concat([results_df, evaluate_model(best_lr, X_test, y_test, 'LR Tuned')], ignore_index=True)
```

```python
dt_pipe = Pipeline(steps=[('preprocessor', preprocessor),
                          ('classifier', DecisionTreeClassifier(random_state=42))])

param_grid_dt = {
    'classifier__criterion': ['gini', 'entropy'],
    'classifier__max_depth': [3, 5, 7, 10, None],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4]
}

grid_dt = GridSearchCV(dt_pipe, param_grid_dt, cv=5, scoring='f1', n_jobs=-1)
grid_dt.fit(X_train, y_train)

print(f"Best DT Parameters: {grid_dt.best_params_}")
best_dt = grid_dt.best_estimator_
results_df = pd.concat([results_df, evaluate_model(best_dt, X_test, y_test, 'DT Tuned')], ignore_index=True)


importances = best_dt.named_steps['classifier'].feature_importances_
```

```python

rf_pipe = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', RandomForestClassifier(random_state=42))])

param_grid_rf = {
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [5, 10, None],
    'classifier__min_samples_leaf': [1, 2]
}
grid_rf = GridSearchCV(rf_pipe, param_grid_rf, cv=3, scoring='f1', n_jobs=-1) # cv=3 for speed
grid_rf.fit(X_train, y_train)
best_rf = grid_rf.best_estimator_
results_df = pd.concat([results_df, evaluate_model(best_rf, X_test, y_test, 'RF Tuned')], ignore_index=True)


ada_pipe = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', AdaBoostClassifier(random_state=42))])
param_grid_ada = {
    'classifier__n_estimators': [50, 100, 200],
    'classifier__learning_rate': [0.01, 0.1, 1.0]
}
grid_ada = GridSearchCV(ada_pipe, param_grid_ada, cv=3, scoring='f1', n_jobs=-1)
grid_ada.fit(X_train, y_train)
best_ada = grid_ada.best_estimator_
results_df = pd.concat([results_df, evaluate_model(best_ada, X_test, y_test, 'AdaBoost Tuned')], ignore_index=True)
```
